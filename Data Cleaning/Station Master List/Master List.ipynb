```python
temp = master_list11.drop(['latitude','longitude','dpcapacity','landmark','id'],axis=1)
temp
```

```python
# now we will begin to narrow down the 0 values, by building a dictionary that contains sets of all of the locations
# where each street is found. Basically, if the set length is equal to one, it means that the values are definitely located
# in the respective city. After this, will manually search the remaining entries with city values == 0 and non-definite city
# locations
temp[['name1','name2']] = temp.name.str.split(" & ",expand=True)
temp = temp.drop('name',axis=1)
```

```python
x = temp[temp['city'] != 0].values
x
```

```python
# Based on first set of names

name_city = {}
name1 = temp['name1'].values
city = temp['city'].values

for i in range(len(name1)):
    if name1[i] not in name_city:
        name_city[f'{name1[i]}'] = []
        name_city[f'{name1[i]}'].append(city[i])
    else:
        name_city[f'{name1[i]}'].append(city[i])
display(name_city)
print(len(name_city))
```

```python
name2 = temp['name2'].values

for i in range(len(name2)):
    
    if name2[i] not in name_city:
        name_city[f'{name2[i]}'] = []
        name_city[f'{name2[i]}'].append(city[i])
    else:
        name_city[f'{name2[i]}'].append(city[i])
display(name_city)
print(len(name_city))
```

```python
set_list = list(name_city.keys())
```

```python
for i in range(len(set_list)):
    name_city[f'{set_list[i]}'] = list(set(name_city[f'{set_list[i]}']))
```

```python
# Now that we have our list, we can sort our actual intersections between roads based on our dictionary. If an intersection
# between two streets have both streets with listed locations containing only Chicago, this means (most likely) that
# the intersection between the two streets is also located in Chicago.

assumed_city = []
index_pos_of_conflicts = []
groupings_of_conflicts = []

for i in range(len(temp)):
        city1 = name_city[f'{name1[i]}']
        city2 = name_city[f'{name2[i]}']
        city1_len = len(set(city1))
        city2_len = len(set(city2))
        
        merged = city1 + city2
        
        #print(i)
        #print(name1[i], name2[i])
        #print(set(city1), set(city2))
        #print(city1 + city2)
        
              
        # If there is only one value in both sets, and the set length == 1 when the sets are merged (they are same val)
        if city1_len == city2_len == len(set(merged)) == 1:
            assumed_city.append(city1[0])
            if city1[0] == 0:
                # record their index positions
                index_pos_of_conflicts.append(i)
                groupings_of_conflicts.append(sorted([name1[i],name2[i]]))
        
        # If set lengths are both two
        elif (city1_len and city2_len) < 3:
            key1, key2 = list(set(merged))[0], list(set(merged))[1]
            
            ctk1 = 0
            ctk2 = 0
            for k in range(len(merged)):
                if merged[k] == key1:
                    ctk1 += 1
                else:
                    ctk2 += 1
            
            # picked 2 to represent that the prescence of a city should be valued more than the precense of a 0
            if type(key1) == str:
                if ctk1 + 2 > ctk2:
                    assumed_city.append(key1)
                elif ctk1 + 2 < ctk2:
                    assumed_city.append(key2)
                else:
                    assumed_city.append(0)
                    index_pos_of_conflicts.append(i)
                    groupings_of_conflicts.append(sorted([name1[i],name2[i]]))
            elif type(key1) == int:
                if ctk2 + 2 > ctk1:
                    assumed_city.append(key2)
                elif ctk2 + 2 < ctk1:
                    assumed_city.append(key1)
                else:
                    assumed_city.append(0)
                    index_pos_of_conflicts.append(i)
                    groupings_of_conflicts.append(sorted([name1[i],name2[i]]))
        # all other cases involve inconclusive feedback on what city the intersections are located in, so we will look it up
        # manually
        else:
            assumed_city.append(0)
            index_pos_of_conflicts.append(i)
            groupings_of_conflicts.append(sorted([name1[i],name2[i]]))

print(assumed_city)
print()
print(index_pos_of_conflicts)
print()
print(groupings_of_conflicts)
```
