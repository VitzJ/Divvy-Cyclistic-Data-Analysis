```python
import pandas as pd
```

```python
df = pd.read_csv(f'202201-divvy-tripdata.csv', header=0)
df.shape
```

```python
df_csvs = []
for i in range(1,13):
    if i > 9:
        temp = f'2022{i}-divvy-tripdata.csv'
    else:
        temp = f'20220{i}-divvy-tripdata.csv'
    df_csvs.append(temp)
        
        
        
print(df_csvs)
print(len(df_csvs))
```

```python
for i, dataframe in enumerate(df_csvs):
    df_csvs[i] = pd.read_csv(dataframe)
```

```python
df_csvs
```

```python
df = pd.concat(df_csvs,ignore_index=True)
```

```python
df.dtypes
```

```python
df = df.rename(columns = {'started_at':'start_time','ended_at':'end_time','start_station_name':'from_station_name', 'start_station_id':'from_station_id','end_station_name':'to_station_name','member_casual':'usertype'})
```

```python
df.to_csv('2022.csv')
```
