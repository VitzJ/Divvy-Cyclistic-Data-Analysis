```python
# I now want to check for repeated values. What I mean is this: I want to build a leger of station ids that corresponds to all
# of the station names currently in the dataframe. I have to check to make sure there are no station names that have multiple
# assignments.

start_station_names = merged2013_2022['start_station_name'].values
start_station_ids = merged2013_2022['start_station_id'].values

end_station_names = merged2013_2022['end_station_name'].values
end_station_ids = merged2013_2022['end_station_id'].values
```

```python
full_dict = {}

for i in range(len(start_station_names)):
    try:
        full_dict[f'{start_station_names[i]}'].append(start_station_ids[i])
    except:
        full_dict[f'{start_station_names[i]}'] = []
        full_dict[f'{start_station_names[i]}'].append(start_station_ids[i])
        pass
    try:
        full_dict[f'{end_station_names[i]}'].append(end_station_ids[i])
    except:
        full_dict[f'{end_station_names[i]}'] = []
        full_dict[f'{end_station_names[i]}'].append(end_station_ids[i])
        pass
```

```python
full_keys = list(full_dict.keys())
```

```python
# This function converts to a set, back to a list, iterates through the list to change all strings to ints, and then converts
# to a set and finally a list again so that all the elements are unique and each dictionary entry is iterable

for i in range(len(full_keys)):
    full_dict[f'{full_keys[i]}'] = list(set(full_dict[f'{full_keys[i]}']))
    
    for k in range(len(full_dict[f'{full_keys[i]}'])):
        try:
            full_dict[f'{full_keys[i]}'][k] = int(float(full_dict[f'{full_keys[i]}'][k]))
        except:
            pass
    full_dict[f'{full_keys[i]}'] = list(set(full_dict[f'{full_keys[i]}']))
```

```python
# I will start off by iterating through the fix_dict and merging together all (Temp), (*) streets

def fix_values(key):
    
    x = fix_dict[str(key)][0]
    
    for i in range(len(x)):

        if x[i][-4:] == ' (*)':
            x[i] = x[i][:-4]
        elif x[i][-7:] == ' (Temp)':
            x[i] = x[i][:-7]
    
    # print(set(x))
    x = list(set(x))
    
    return x
```

```python
def fix_id(change_to_id=int, change_to_name='', edit_list=[]):
    for i in range(len(edit_list)):
        if change_to_name != edit_list[i]:
        
            merged2013_2022.loc[merged2013_2022['start_station_name'] == edit_list[i], 'start_station_name'] = change_to_name
            merged2013_2022.loc[merged2013_2022['end_station_name'] == edit_list[i], 'end_station_name'] = change_to_name
        
        merged2013_2022.loc[merged2013_2022['start_station_name'] == change_to_name, 'start_station_id'] = change_to_id
        merged2013_2022.loc[merged2013_2022['end_station_name'] == change_to_name, 'end_station_id'] = change_to_id
```
