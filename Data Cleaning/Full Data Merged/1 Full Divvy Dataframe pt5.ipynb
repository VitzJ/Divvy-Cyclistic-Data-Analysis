```python
start_sta = merged2013_2022['start_station_name'].unique()
end_sta = merged2013_2022['end_station_name'].unique()
print(f'Start_stations: {start_sta}')
print(f'End_stations: {end_sta}')
```

```python
len_ct1 = 0
name_list1 = []
index1 = []
for i in range(len(start_sta)):
    if start_sta[i] not in end_sta:
        name_list1.append(start_sta[i])
        k = merged2013_2022[merged2013_2022['start_station_name'] == start_sta[i]]
        len_ct1 += len(k)
        index1 += list(k.index)
        
print(len_ct1)
```

```python
len_ct2 = 0
name_list2 = []
index2 = []
for i in range(len(end_sta)):
    if end_sta[i] not in start_sta:
        name_list2.append(end_sta[i])
        k = merged2013_2022[merged2013_2022['end_station_name'] == end_sta[i]]
        len_ct2 += len(k)
        index2 += list(k.index)
print(len_ct2)
```

```python
# Fix landmarks

start_sta = list(merged2013_2022['start_station_name'].unique())
end_sta = list(merged2013_2022['end_station_name'].unique())
```

```python
land_dict = {}

for i in range(len(start_sta)):
    land_dict[f'{start_sta[i]}'] = list(merged2013_2022.loc[merged2013_2022['start_station_name'] == start_sta[i]]['start_landmark'].unique())
```

```python
for i in range(len(end_sta)):
    land_dict[f'{end_sta[i]}'] += list(merged2013_2022.loc[merged2013_2022['end_station_name'] == end_sta[i]]['end_landmark'].unique())
```

```python
land_dict
```

```python
sorted(end_sta) == sorted(start_sta)
```

```python
for i in range(len(end_sta)):
    land_dict[f'{end_sta[i]}'] = sorted(list(set(land_dict[f'{end_sta[i]}'])), reverse=True)
```

```python
land_dict
```

```python
fix_list = list(land_dict.keys())
```

```python
for i in range(len(fix_list)):
    temp = land_dict[fix_list[i]]
    if len(temp) > 1:
        merged2013_2022.loc[merged2013_2022['start_station_name'] == fix_list[i], 'start_landmark'] = temp[0]
        merged2013_2022.loc[merged2013_2022['end_station_name'] == fix_list[i], 'end_landmark'] = temp[0]
```
