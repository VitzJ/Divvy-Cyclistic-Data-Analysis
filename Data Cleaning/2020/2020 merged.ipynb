```python
import pandas as pd
```

```python
df = pd.read_csv('Divvy_Trips_2020_Q1.csv', header=0)
df.shape
```

```python
for i in range(4,13):
    if i > 9:
        temp = pd.read_csv(f'2020{i}-divvy-tripdata.csv')
        print(temp.shape)
    else:
        temp = pd.read_csv(f'20200{i}-divvy-tripdata.csv')
        print(temp.shape)
    df = pd.concat([df, temp],ignore_index=True)
```

```python
df
```

```python
df = df.rename(columns = {'started_at':'start_time','ended_at':'end_time','start_station_name':'from_station_name', 'start_station_id':'from_station_id','end_station_name':'to_station_name','member_casual':'usertype'})
```
